```python
def overlap_graph(fasta, n):
    results = []
    for key1 in fasta:
        for key2 in fasta:
            if key1 != key2 and fasta[key1].endswith(fasta[key2][:n]):
                results.append((key1, key2))

    return results

fo = open("rosalind_grph.txt")
segments = {} #segments : our dict with fasta code as key and seq as it's value
#read from file so that if line begins with >, store as key and all following as value for that key

for line in fo:
    line = line.rstrip()
    if line[0] == '>':
        segments[line.strip('>')] = ""
        temp = line.strip('>')
    else:
        segments[temp] += line

        
for edge in overlap_graph(segments,3):
    print(edge[0], edge[1])
    
fo.close()
```
